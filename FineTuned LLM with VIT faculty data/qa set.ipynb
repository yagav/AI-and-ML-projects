```python
import pandas as pd
```

```python
ds = pd.read_csv('vit faculty.csv')
```

```python
ds.head()
```

```python
question =[]
answer = []
for i in range(18):
    question.append(f"What is the email id of {ds.iloc[i]['Faculty Name']}")
    answer.append(f"{ds.iloc[i]['Contact Email']}")
    question.append(f"What are the Publications of {ds.iloc[i]['Faculty Name']}")
    answer.append(f"{ds.iloc[i]['Publications']}")
    question.append(f"What are the reseach areas of {ds.iloc[i]['Faculty Name']}")
    answer.append(f"{ds.iloc[i]['Research Interests']}")
    question.append(f"What are the qualifications of the faculty {ds.iloc[i]['Faculty Name']}")
    answer.append(f"{ds.iloc[i]['Qualification']}")
    question.append(f"What are the courses taught by {ds.iloc[i]['Faculty Name']}")
    answer.append(f"{ds.iloc[i]['Courses Taught']}")

print(question[5])
print(answer[5])
    
```

```python
js=[]

for q,a in zip(question, answer):
    js.append({
    "conversation": [
        {"role":"user","content":q},
        {"role":"assistant","content":a}
    ]
})
        

```
